```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import matplotlib.patches as mpatches
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
ALIGNED_DIR = "/workspaces/Gesture-Language-Alignment-in-Instructional-Videos/00_videos/aligned_transcripts"
all_data = []

for file in os.listdir(ALIGNED_DIR):
    if file.endswith("_aligned.json"):
        df = pd.read_json(os.path.join(ALIGNED_DIR, file))
        df["video_name"] = file.replace("_aligned.json", "")
        all_data.append(df)

df_all = pd.concat(all_data, ignore_index=True)

plt.figure(figsize=(6, 4))
sns.boxplot(x="has_imperative", y="gesture_avg", data=df_all)
plt.xticks([0, 1], ["No Imperative", "Has Imperative"])
plt.title("Gesture Intensity by Imperative Presence")
plt.xlabel("Imperative")
plt.ylabel("Average Gesture Motion")
plt.tight_layout()
plt.savefig("/workspaces/Gesture-Language-Alignment-in-Instructional-Videos/assets/gesture_by_has_imperative.png")

plt.figure(figsize=(6, 4))
sns.boxplot(x="imperative_type", y="gesture_avg", data=df_all)
plt.title("Gesture Intensity by Imperative Type")
plt.xlabel("Imperative Type")
plt.ylabel("Average Gesture Motion")
plt.tight_layout()
plt.savefig("/workspaces/Gesture-Language-Alignment-in-Instructional-Videos/assets/gesture_by_imperative_type.png")
```
